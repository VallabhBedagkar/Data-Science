```python
X = df['TV'].values.reshape(-1, 1)
y = df['Sales']
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2,random_state = 0)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.linear_model import LinearRegression
Linear_reg = LinearRegression()
Linear_reg.fit(X_train,y_train)
```

```python
Linear_reg.coef_
```

```python
Linear_reg.intercept_
```

```python
y_pred = Linear_reg.predict(X_train)
y_pred
```

```python
plt.scatter(X_train , y_train , color = 'teal')
plt.plot(X_train , y_pred,color = 'coral')
plt.title("TV AD Vs Sales")
plt.xlabel ('Expendicture on TV AD')
plt.ylabel ('Sales')
```

```python
from sklearn.metrics import mean_squared_error, r2_score
r_squared = r2_score(y_train, y_pred)
mse = mean_squared_error(y_train, y_pred)
print("Linear Regression Summary:")
print(f"Coefficients: {Linear_reg.coef_}")
print(f"Intercept: {Linear_reg.intercept_}")
print(f"R-squared: {r_squared}")
print(f"Mean Squared Error: {mse}")
```
